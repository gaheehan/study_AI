## 기계 학습

머신 러닝(ML)은 컴퓨터가 데이터 변수 간의 패턴과 관계를 학습하는 능력을 말한다.

## 1. 지도 학습 기법

지도 학습 기법에는 라벨링된 훈련 데이터가 필요하다. 예를 들어, 유튜브 동영상에 좋아요를 누르거나, 평가하지 않거나, 싫어한다는 라벨이 지정될 수 있다. 유튜브의 추천 모델은 학습할 때 필요한 훈련 데이터의 라벨이다. 훈련 데이터가 많을 수록 더 정확한 모델을 만들 수 있다.

라벨이 적절해야한다. 예를들어 공부한 시간이 주어졌을 때 학생의 시험 점수를 예측하려면 데이터에 각 학생의 시험 점수가 포함되어야 한다. 이렇게 하면 기계 학습 알고리즘이 주어진 예로부터 학습할 수 있다.

지도 학습 기술의 또 다른 중요한 용어는 '특성(feature)'이다. 특성은 대상 또는 라벨을 예측하는데 사용할 수 있는 데이터를 나타낸다. 시험 점수 예시에서는 시험 점수가 대상이며, 공부한 시간이 특성이다. 시험 점수 예에서는 공부한 문제 수, 시험 전 수면 시간 등이 또 다른 특성이 될 수 있다.

특성 및 라벨/대상을 사용하여 알고리즘은 특성과 라벨/대상 간의 관계를 "학습"할 수 있다.

지도 학습 기술은 데이터를 다른 그룹으로 분류하거나 변수 간의 관계를 예측하는 데 사용할 수 있다. 데이터가 속한 범주 또는 그룹을 분류하는 기술이다. 예를 들어, 내일 비가 올지 여부를 예측하려면 비가 오거나 맑음과 같은 범주를 반환하는 알고리즘을 사용할 수 있다. 반면 회귀 분석에 사용되는 기술은 수치 데이터를 반환한다. 예를 들어 내일 비의 양을 예측하려면 범주 대신 숫자 데이터를 반환해야 한다.

분류와 회귀의 차이점

분류: 그룹 또는 클래스의 예측이 필요한 모든 시나리오 예:만조 대 간조, 승패 등


회기: 숫자 예측이 필요한 모든 시나리오 예:사람의 몸무게,판매금액 등

# 1.1 K-최근접 이웃 알고리즘(K-Nearest Neighbor)

<font color=red>K-최근접 이웃(K-NN, K-Nearest Neighbor) 알고리즘</font>은 가장 간단한 머신러닝 알고리즘으로, 분류(Classification) 알고리즘이다. 비슷한 특성을 가진 데이터는 비슷한 범주에 속하는 경향이 있다는 가정하에 사용한다.

image.png

예를 들어 위와 같이 데이터가 주어져 있을 때, 빨간색인 세모 모양의 데이터는 초록색 그룹과 노란색 그룹 중 어디에 속한다고 말할 수 있을까?
주변에 가까운 데이터들이 모두 노란색이기 때문에 '노란색 그룹에 속할 것이다'라고 추측할 수 있다.
이처럼, <font color=blue> 주변의 가장 가까운 K개의 데이터를 보고 데이터가 속할 그룹을 판단하는 알고리즘이 K-NN 알고리즘이다.</font>

K-NN 알고리즘은, 단순히 훈련 데이터셋을 그냥 저장하는 것이 모델을 만드는 과정의 전부이다. 그리고 거리를 측정할 땐 유클리드 거리(Euclidean distance)를 사용한다.
K-NN 알고리즘의 특징 중 하나는 **K의 값에 따라 분류가 달라질 수 있다는 점**이다.
예를 들어 아래와 같은 상황을 보면 K = 1인 경우에는 초록색 그룹이라고 판단하고, K = 3인 경우에는 노란색 그룹이라고 판단할 것이다.

image.png

그리고, 항상 분류가 가능하도록 K는 홀수로 설정하는 것이 좋으며, 최선의 K를 선택하는 것은 데이터마다 다르게 접근해야 하는데, 일반적으로는 총 데이터 수의 제곱근 값을 사용한다.

K-NN 알고리즘은 너무나 간단한 알고리즘이지만, 실제로 이미지 처리, 글자/얼굴 인식, 추천 알고리즘, 의료 분야 등에서 많이 사용된다. 그렇다면 장단점은 뭘까?


장점: **단순하기 때문에 다른 알고리즘에 비해 구현하기 쉽다** 
훈련 데이터를 그대로 가지고 있어 특별한 훈련을 하지 않기 때문에 **훈련 단계가 매우 빠르게 수행된다.**

단점: 모델을 생성하지 않기 때문에 **특징과 클래스 간 관계를 이해하는데 제한적이다.**
모델의 결과를 가지고 해석하는 것이 아니라, 미리 변수와 클래스 간의 관계를 파악하여 이를 알고리즘에 적용해야 원하는 결과를 얻을 수 있기 때문

또, **적절한 K의 선택이 필요**하고, 훈련 단계가 빠른 대신 **데이터가 많아지면 분류 단계가 느리다.**

### 라벨 인코딩

라벨 인코딩은 각 클래스 범주에 번호를 할당하는 것을 의미. 예를 들어, 날씨 예측의 경우 비와 맑음 두가지 클래스가 있는 경우, 비는 0, 맑음은 1로 라벨을 지정할 수 있다. 

Iris Flower 데이터 세트의 클래스(꽃 유형)를 숫자로 변환할 수 있습니다. 데이터 세트 내의 클래스는 무엇이 있습니까? 총 몇 개의 클래스가 있습니까?

3개
 'Iris-setosa'는 1, 'Iris-versicolor'는 2, 'Iris-virginica'는 3

이제 iris 데이터 세트에 클래스를 라벨 인코딩을 합니다. 'Iris-setosa'는 1, 'Iris-versicolor'는 2, 'Iris-virginica'는 3이 될 수 있습니다. 아래는 클래스 중 하나에 대해 라벨 인코딩한 인코딩한 코드입니다. 모든 클래스에 대해 라벨 인코딩을 진행하는 코드로 편집해보세요.

```python
# 각 클래스의 데이터 포인트 수 출력:
print(df['class'].value_counts())

# 클래스에 각 숫자를 지정하는 딕셔너리
label_encode = {"class": {"Iris-setosa":1,  'Iris-versicolor':2,'Iris-virginica':3}}

# .replace를 사용하여 다른 클래스를 숫자로 변경
df.replace(label_encode, inplace=True)    # 원본 DataFrame을 치환하고 싶은 경우 inplace=True를 설정

# 각 클래스의 데이터 포인트 수를 출력하여 클래스가 숫자로 변경되었는지 확인
print(df['class'].value_counts())
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

k의 기본 값 = 5 이다.

처음에는 2가지 특성(sepal_length 및 sepal_width)만 사용합니다. 아래 코드에서 이들은 'x' 데이터 프레임에 포함됩니다. 꽃의 '클래스', 즉 꽃 종의 이름은 'y' 데이터 프레임에 포함된 '라벨'이 됩니다.

```python
# KNeighborsClassifier 초기화
KNN = KNeighborsClassifier()

# X 값과 y 값을 추출. x는 특성이고 y는 클래스이다.
X = df[['sepal_length','sepal_width']]
y = df['class']

# 데이터가 올바른지 확인하기 위해 X 및 y의 .head()를 출력한다.
print(X.head())
print(y.head())

# X 및 y 값을 사용하여 KNN 훈련시킨다. 훈련은 fit 메소드를 통해 수행
knn = KNN.fit(X,y)

# sepal_length = 5 및 sepal_width = 3 인 경우 학습된 KNN을 사용하여 꽃의 유형을 예측
# .predict 메서드 사용

test = pd.DataFrame()
test['sepal_length'] = [5]
test['sepal_width'] = [3]
predict_flower = KNN.predict(test)

# predict_flower 출력
print(predict_flower)
```

위의 출력에서 sepal_length = 5, sepal_width = 3인 데이터 포인트에 대해 어떤 유형의 꽃이 예측되었습니까?

## 의사결정트리(Decision Tree)란? 
[참조](https://wooono.tistory.com/104)
- 의사결정트리는 일련의 분류 규칙을 통해 데이터를 분류, 회귀하는 지도 학습 모델 중 하나

image.png

- 위 그림은 대표적인 의사결정트리의 예시로서, 타이타닉호의 탑승객의 생존여부를 나타낸다.

- 이렇게 특정 기준(질문)에 따라 데이터를 구분하는 모델을 의사 결정 트리 모델이라고 함

- 한번의 분기 때마다 변수 영역을 두 개로 구분

- 결정 트리에서 질문이나 정답은 노드(Node)라고 불림.
 - 맨 처음 분류 기준: Root Node
 - 중간 분류 기준: Intermediate Node
 - 맨 마지막 노드: Terminal Node 혹은 Leaf Node
 - 결정 트리의 기본 아이디어는, Leaf Node가 가장 섞이지 않은 상태로 완전히 분류되는 것, 즉 복잡성(entropy)이 낮도록 만드는 것이다.

 


scikit learn에서 의사 결정 트리 가져오기

```python
df3 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/review/17/iris.data", header=None)
names = ["sepal_length", "sepal_width", "petal_length", "petal_width","class"]
df3.columns = names
df3.head()
```

```python
from sklearn import tree
```

```python
# 의사 결정 트리 초기화
dt = tree.DecisionTreeClassifier()

# X,y 값 추출 X는 특성이고 y는 클래스이다.
X = df3[['sepal_length','sepal_width']]
y = df3['class']

# 데이터가 올바른지 확인하기 위해 X 및 y의 .head()를 출력
print(X.head())
print(y.head())

# 의사결정 트리 훈련
DT = dt.fit(X,y)
```
